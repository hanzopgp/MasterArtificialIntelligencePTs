```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.naive_bayes as nb
from sklearn import svm
from sklearn import linear_model as lin
from sklearn.model_selection import train_test_split
import string
import re
import unicodedata
import nltk
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

import codecs
import re
import os.path
# Chargement des données:
def load_pres(fname):
    alltxts = []
    alllabs = []
    s=codecs.open(fname, 'r','utf-8') # pour régler le codage
    while True:
        txt = s.readline()
        if(len(txt))<5:
            break
        lab = re.sub(r"<[0-9]*:[0-9]*:(.)>.*","\\1",txt)
        txt = re.sub(r"<[0-9]*:[0-9]*:.>(.*)","\\1",txt)
        print(lab)
        if lab.count('M') >0:
            alllabs.append(-1)
        else: 
            alllabs.append(1)
        alltxts.append(txt)
    return alltxts,alllabs

import sklearn.naive_bayes as nb
from sklearn.metrics import f1_score,accuracy_score,confusion_matrix,classification_report,roc_auc_score
fname = "data/corpus.tache1.learn.utf8"
# alltxts,alllabs = load_pres(fname)
X_train ,y_train = np.array(alltxts),np.array(alllabs)
fname_test = "data/corpus.tache1.test.utf8"
alltxts_test, alllabs_test = load_pres(fname_test)
X_test = np.array(alltxts_test)
y_test = np.array(alllabs_test)
print(y_test)
stop_word_fr = stopwords.words("french")
vectorizer = CountVectorizer(analyzer='word',stop_words=stop_word_fr,strip_accents='ascii',lowercase=True,ngram_range=(1,1),min_df=0.,max_df=1.)
X_train_2 = vectorizer.fit_transform(X_train,y_train)
X_test_2 = vectorizer.transform(X_test)
#print(vectorizer.get_feature_names())
clf = nb.MultinomialNB()
X_train_3 = clf.fit(X_train_2,y_train)
#X_test_3 = clf.transform(X_test_2)
y_pred_3 = clf.predict(X_test_2)
print("classifieur:",accuracy_score(y_test,y_pred_3)) # pour rentrer dans le classifieur... Depend évidemment du classifieur!
print("classifieur:",f1_score(y_test,y_pred_3)) # pour rentrer dans le classifieur... Depend évidemment du classifieur!
print("classifieur:",roc_auc_score(y_test,y_pred_3)) # pour rentrer dans le classifieur... Depend évidemment du classifieur!
confusion_matrix(y_test,y_pred_3)
print(classification_report(y_test,y_pred_3)) 
import sklearn.naive_bayes as nb
from sklearn.metrics import f1_score,accuracy_score,confusion_matrix,classification_report,roc_auc_score
```
