```python
import gym
from stable_baselines3 import PPO
```

```python
env = gym.make("gym_rocketlander:rocketlander-v0")
n_steps = 3_000_000

model = PPO(
  env = env,
  policy= 'MlpPolicy',
  n_steps= 1024,
  batch_size= 64,
  gae_lambda= 0.98,
  gamma= 0.999,
  n_epochs= 4,
  ent_coef= 0.01,
  verbose=1
)

model.learn(total_timesteps=n_steps)
# model.save("/content/drive/MyDrive/ppo_discret")
```
