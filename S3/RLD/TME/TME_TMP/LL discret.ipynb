```python
import gym
from stable_baselines3 import PPO
```

```python
env = gym.make("LunarLander-v2")

model = PPO(
  # n_envs= 16,
  env = env,
  # n_timesteps= 1e6,
  policy= 'MlpPolicy',
  n_steps= 1024,
  batch_size= 64,
  gae_lambda= 0.98,
  gamma= 0.999,
  n_epochs= 4,
  ent_coef= 0.01,
  verbose=1
)

model.learn(total_timesteps=50_000_000)
model.save("/content/drive/MyDrive/ppo_discret")
```
